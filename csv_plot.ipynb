### How to plot a Pandas Dataframe with Matplotlib?
https://www.geeksforgeeks.org/how-to-plot-a-pandas-dataframe-with-matplotlib/

```python
import pandas as pd # Data Analysis
import yfinance as yf # Yahoo Finance, where the data comes from
import matplotlib.pyplot as plt # Data Visualization
```

```python
stock = "AAPL"
stock_ticker = yf.Ticker(stock)
df_history = stock_ticker.history(period="max")        # Dataframe   FROM   Yahoo Finance
df_history.to_csv(f'{stock} csv.csv')                  # Dataframe    TO    CSV
df_history = pd.read_csv(f'{stock} csv.csv')           # Dataframe   FROM   CSV
df_history.plot(x="Date", y="Close", kind="line", title=stock_ticker.info['shortName'],figsize=(16,9))
```
