# [Bokeh: Plotting Stock Data with a Range Tool](https://www.youtube.com/watch?v=zSTLzp-4EUg)

```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.io import output_notebook
output_notebook()   # needed for Notebook in-cell plots vs html
```

```python
# Data
data = yf.download('^GSPC')         #'^GSPC'
data.reset_index(inplace=True)
# print(data.head())
```

```python
dates = np.array(data["Date"])
# print(dates)
```

```python
source = ColumnDataSource(data=data)
```

```python
p = figure(plot_height=300, plot_width=1200, tools="xpan",
	toolbar_location=None, x_axis_type="datetime", x_axis_location="above",
		x_range=(dates[1500], dates[2500]))
```

```python
p.line('Date','Adj Close', source=source)
p.yaxis.axis_label = "Price"
select = figure(title="Drag the middle and edges of the Selection Box to Change the Range",
	plot_height=150, plot_width=1200, y_range=p.y_range, x_axis_type="datetime",y_axis_type=None,
	tools="", toolbar_location=None)
#p.x_range
```

```python
range_tool = RangeTool(x_range=p.x_range)
range_tool.overlay.fill_color="navy"
range_tool.overlay.fill_alpha = 0.2
```

```python
select.line('Date', 'Adj Close', source=source)
select.ygrid.gird_line_color=None
select.add_tools(range_tool)
select.toolbar.active_multi = range_tool
```

```python
show(column(p, select))
```
