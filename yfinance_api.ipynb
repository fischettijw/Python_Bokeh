```python
import yfinance as yf

# provide stock ticker symbol
stock= yf.Ticker("^GSPC")

# save stock history (Date, Open, High, LOW, Close, Volume)
stock_history = stock.history(start="1985-01-01", end="2021-10-09", interval="1d")

# save actions (dividends, splits)
stock_actions = stock.actions

# save dividends
stock_dividends = stock.dividends

# save splits
stock_splits = stock.splits

# display columns in stock history as initially created
print('Before Index Reset:',stock_history.columns, '\n')

# change index (Date) into a column
stock_history.reset_index(inplace=True)

# display columns in stock history after changing index(Date) into column
print('After Index Reset:',stock_history.columns, '\n')

# display stock history
print(stock_history)




```
